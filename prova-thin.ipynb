```python
import numpy as np
import scipy.misc
import cv2
import matplotlib.pyplot as plt
import thinning as th
%matplotlib inline
```

```python
#mask = cv2.imread('mask.png')
mask = cv2.imread('Box1.tif')
mask1 = scipy.misc.imresize(mask, 0.25) 
mask_g = cv2.cvtColor(mask1, cv2.COLOR_BGR2GRAY)
ret, mask_bw = cv2.threshold(mask_g, 127, 255, cv2.THRESH_BINARY)
mask_bw = 255 - mask_bw
```

```python
plt.figure(figsize=(16,16))
plt.imshow(mask_bw,cmap='gray');
```

```python
bw2 = th.thinning(mask_bw)
```

```python
plt.figure(figsize=(16,16))
plt.imshow(bw2,cmap='gray');
```

```python
src = cv2.distanceTransform(mask_bw, cv2.DIST_L2, cv2.DIST_MASK_PRECISE)
```

```python
plt.figure(figsize=(16,16))
plt.imshow(src,cmap='gray');
```

```python
#blur = cv2.bilateralFilter(src,9,75,75)
k = 15
kernel = np.ones((k,k),np.float32)/k**2
blur = cv2.filter2D(src,-1,kernel)
```

```python
plt.figure(figsize=(16,16))
plt.imshow(blur,cmap='gray');
```

```python
blurred = True

if blurred:
    img = blur
else:
    img = src
```
